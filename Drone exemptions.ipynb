```python
## Count categories better
## Since the categories are comma-separated lists, let's explode

## turn a column of comma-separated lists into a list of discrete values
def discrete_values (col):
    categories = []
    map(lambda x : (map(lambda y : categories.append(y.strip().lower()), x.split(","))), col)
    disc = pd.Series(categories).to_frame()[0].value_counts().to_frame()
    disc.reset_index(level=0, inplace=True)
    disc.columns=["category","count"]
    return disc
             
discrete_values(ctexemptions["category"])

## Variable for discrete categories
#categories = []
#map(lambda x : (map(lambda y : categories.append(y.strip().lower()), x.split(","))), ctexemptions["category"])

#discrete_categories = pd.Series(categories).to_frame()[0].value_counts().to_frame()
#discrete_categories.reset_index(level=0, inplace=True)
#discrete_categories.columns=["category","exemptions"]
                             
#discrete_categories
```
